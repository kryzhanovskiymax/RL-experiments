```python
from env import CardGame
from agent import Agent
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def draw_heatmaps(stats):
    plt.figure(figsize=(30, 10))
    sns.heatmap(stats, annot=True, fmt=".1f", cmap="YlGnBu", cbar_kws={'label': 'Percentage of Pushes'})
    plt.title('Small Blind Push Stats')
    plt.xlabel('Effective Stack')
    plt.ylabel('Card Value (0 to 12)')
    plt.tight_layout()
    plt.show()
```

```python
agent1 = Agent()
agent2 = Agent()
game = CardGame(agent1=agent1,
                agent2=agent2)
```

```python
game.play(5000000000000)
bb_stats, sb_stats = game.get_play_stats()
```

```python
draw_heatmaps(sb_stats)
```

```python
draw_heatmaps(bb_stats)
```
